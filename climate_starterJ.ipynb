# Exploratory Precipitation Analysis

```python
# check: 
# Find the most recent date in the data set (note that the last day in the dataset is 8/23/2017)
# Creates a pandas DataFrame using the date and precipitation columns
# Sorts the DataFrame by date
# Makes a plot using pandas with date as the x and precipitation as the y variables
```

```python
# get a list of columns and types if data in measurement table
inspector = inspect(engine)
columns = inspector.get_columns('measurement')
for c in columns:
    print(c['name'], c["type"])
```

```python
# Find the most recent date in the data set (note that the last day in the dataset is 8/23/2017)
latest_date = session.query(Measurement.date).order_by(Measurement.date.desc()).first()
print(latest_date)
```

```python
##ERROR:type object 'measurement' has no attribute 'data' 
#try to access an attribute of that returned object
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results. 
# querying the 12 preceding months of data-- datatime as dt

# Starting from the most recent data point in the database
# Calculate the date one year from the last date in data set.
recentDate = dt.date(2017,8,23)-dt.timedelta(days=365)

# Perform a query to retrieve the data and precipitation scores- filter
precipitDF = session.query(Measurement.date, Measurement.prcp).filter(Measurement.date> recentDate).all()
print(recentDate, precipitDF)
```

```python
# Save the query results as a Pandas DataFrame and 
#set the index to the date column-- change function inplace = False to True
precipitDF = pd.DataFrame(precipitDF, columns =['date','prcp'])
# set index to the date
precipitDF.set_index('date', inplace=True)
precipitDF
```

```python
# Sort the dataframe by date
precipitDF.sort_values(by ='date')
```

```python
# Use Pandas Plotting with Matplotlib to plot the data
precipitDF.plot(rot=45)
plt.xlabel("Date")
plt.ylabel("inches")
plt.title("precipitation")
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data

#df = pd.DataFrame(precipitDF)
#df.describe()

precipitDF.describe(include = 'all')
```
